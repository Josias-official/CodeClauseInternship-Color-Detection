```python
# Import necessary libraries
import cv2  # OpenCV for image processing
import pandas as pd  # Pandas for data manipulation

# Define the path to the image file
path = r'color.jpg'
img = cv2.imread(path)  # Read the image using OpenCV

# Declare global variables (used later on)
clicked = False  # Flag to track mouse clicks
r = g = b = x_pos = y_pos = 0  # Initialize color and position variables

# Read a CSV file containing color information using Pandas
index = ["color", "color_name", "hex", "R", "G", "B"]
csv = pd.read_csv('colors.csv', names=index, header=None)

# Define a function to calculate the closest matching color
def get_color_name(R, G, B):
    minimum = 10000
    for i in range(len(csv)):
        d = abs(R - int(csv.loc[i, "R"])) + abs(G - int(csv.loc[i, "G"])) + abs(B - int(csv.loc[i, "B"]))
        if d <= minimum:
            minimum = d
            cname = csv.loc[i, "color_name"]
    return cname

# Define a function to capture mouse double clicks and get pixel color
def draw_function(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        global b, g, r, x_pos, y_pos, clicked
        clicked = True
        x_pos = x
        y_pos = y
        b, g, r = img[y, x]  # Get the color values at the clicked position

        # Convert color values to integers
        b = int(b)
        g = int(g)
        r = int(r)

# Create a window for the image and set the mouse callback function
cv2.namedWindow('image')
cv2.setMouseCallback('image', draw_function)

# Main loop for displaying and interacting with the image
while True:
    cv2.imshow("image", img)  # Display the image

    if clicked:
        # Draw a rectangle with the selected color at the top of the image
        cv2.rectangle(img, (20, 20), (750, 60), (b, g, r), -1)

        # Create a text string displaying the color name and RGB values
        text = get_color_name(r, g, b) + ' R=' + str(r) + ' G=' + str(g) + ' B=' + str(b)

        # Display the text on the image
        cv2.putText(img, text, (50, 50), 2, 0.8, (255, 255, 255), 2, cv2.LINE_AA)

        # For very light colors, display text in black
        if r + g + b >= 600:
            cv2.putText(img, text, (50, 50), 2, 0.8, (0, 0, 0), 2, cv2.LINE_AA)

        clicked = False

    # Break the loop when the user presses the 'esc' key
    if cv2.waitKey(20) & 0xFF == 27:
        break

# Close all OpenCV windows
cv2.destroyAllWindows()
```
